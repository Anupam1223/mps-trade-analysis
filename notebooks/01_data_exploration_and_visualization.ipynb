```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dotenv import load_dotenv
import os
import sys
```

```python
load_dotenv()
sys.path.append(os.path.join(os.getcwd(), 'src'))

from data_ingestion import fetch_data

print("Fetching data...")
raw_data_dict = fetch_data()
```

```python
for symbol, df in raw_data_dict.items():
    print(f"\n=== EDA Report for {symbol} ===")

    # Move index to a column if it's datetime
    df = df.copy()
    df.reset_index(inplace=True)

    # Rename index column to 'timestamp' (optional but recommended)
    if 'index' in df.columns:
        df.rename(columns={'index': 'timestamp'}, inplace=True)

    # Make sure it's datetime
    if not pd.api.types.is_datetime64_any_dtype(df['timestamp']):
        df['timestamp'] = pd.to_datetime(df['timestamp'])

    # Sort by time
    df.sort_values(by='timestamp', inplace=True)

    # Basic Info
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print("Data Types:\n", df.dtypes)
    print("Missing Values:\n", df.isnull().sum())
    print("Time Range:", df['timestamp'].min(), "to", df['timestamp'].max())
    print("\nSample Data:\n", df.head())
    print("\nDescriptive Statistics:\n", df.describe())

    # --- Plot Closing Price Over Time ---
    plt.figure(figsize=(12, 4))
    plt.plot(df['timestamp'], df['close'], label='Close Price', color='blue')
    plt.title(f'{symbol} Closing Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # --- Rolling Stats ---
    rolling_mean = df['close'].rolling(window=30).mean()
    rolling_std = df['close'].rolling(window=30).std()

    plt.figure(figsize=(12, 4))
    plt.plot(df['close'], label='Close', color='blue')
    plt.plot(rolling_mean, label='30-Day MA', color='red')
    plt.plot(rolling_std, label='30-Day STD', color='black')
    plt.title(f'{symbol} Close Price with Rolling Mean & Std')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # --- Daily Returns ---
    df['daily_return'] = df['close'].pct_change()

    plt.figure(figsize=(12, 4))
    plt.plot(df['timestamp'], df['daily_return'], color='purple')
    plt.title(f'{symbol} Daily Returns Over Time')
    plt.xlabel('Date')
    plt.ylabel('Return')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # --- Distribution Plots ---
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    sns.histplot(df['daily_return'].dropna(), kde=True, ax=ax1, bins=50)
    ax1.set_title(f'{symbol} Daily Returns Distribution')

    sns.histplot(np.log1p(df['volume']), kde=True, ax=ax2, color='green', bins=50)
    ax2.set_title(f'{symbol} Log(1 + Volume) Distribution')

    plt.tight_layout()
    plt.show()
```
